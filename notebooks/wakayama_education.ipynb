```python
# this data is about statistics on education in 2015, fetched from e-stats
data = pd.read_pickle("prefecture_education.pickle") 
```

```python
data["大学相当"] = data.iloc[:,14] + data.iloc[:,16] + data.iloc[:,17] 
```

# Distribution on education per sector

## Distribution on all types of education

```python
data.iloc[1:,3].plot(figsize=[19,6])
pref = data.iloc[1:,2].values
plt.tick_params(labelsize=13)
plt.title("Distribution of Numbers of all types of education in Japan in 2015",size=20)
plt.xticks(list(range(0,47)),pref,rotation='vertical')
plt.ylabel("total numbers of schools",size=18)
plt.show()
```

# z-scores on 和歌山

```python
educations = ["計","幼稚園","小学校","中学校","高等学校","大学相当"]
for col in educations:
    col_zscore = col + '_zscore'
    data[col_zscore] = (data[col] - data[col].mean())/data[col].std(ddof=0)
```
